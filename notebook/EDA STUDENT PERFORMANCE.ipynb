```python
# 3,8 Adding columns for "total score" and "average score"

df['total score'] = df['math score'] + df['reading score'] + df['writing score']
df['average score'] = df['total score'] / 3
df.head()
```

```python
reading_full =df[df['reading score'] == 100]['average score'].count()
writing_full =df[df['writing score'] == 100]['average score'].count()
math_full =df[df['math score'] == 100]['average score'].count()

print(f'Number of students with full score in Reading: {reading_full}')
print(f'Number of students with full score in Writing: {writing_full}')
print(f'Number of students with full score in Math: {math_full}')
```

```python
reading_less_20 = df[df['reading score'] < 20]['average score'].count()
writing_less_20 = df[df['writing score'] < 20]['average score'].count()
math_less_20 = df[df['math score'] < 20]['average score'].count()

print(f'Number of students with less than 20 score in Reading: {reading_less_20}')
print(f'Number of students with less than 20 score in Writing: {writing_less_20}')
print(f'Number of students with less than 20 score in Math: {math_less_20}')
```

```python
# 4.1.1) Histogram & KDE
# Kernel Density Estimation(KDE) is used to plot the distribution of a variable.
fig, axs = plt.subplots(1,2,figsize=(15,7)) # 1 row and 2 columns
plt.subplot(121) # 1st row and 2nd column
sns.histplot(data=df,x='average score',bins=30,kde=True,color='g') # Here we are plotting histogram and kde for average score
plt.title('Distribution of Average Score',fontsize=15)
plt.xlabel('Average Score',fontsize=12)
plt.ylabel('Number of Students',fontsize=12)
plt.subplot(122) # 1st row and 2nd column
sns.histplot(data=df,x='average score',kde=True,hue='gender') # this line is used to plot histogram and kde for average score based on gender
plt.title('Distribution of Average Score',fontsize=15)
plt.xlabel('Average Score',fontsize=12)
plt.ylabel('Number of Students',fontsize=12)
plt.show()
```

### 4.3 Multivariate analysis using pieplot

```python
plt.rcParams['figure.figsize'] = (30,12) # rcparmas to set default figure size

plt.subplot(1,5,1) # plot in 1 row and 5 columns, 1st plot
size = df['gender'].value_counts()
labels = 'Female', 'Male'
color=['red','green']

plt.pie(size, colors = color, labels=labels, autopct='%2f%%')
plt.title('Gender', fontsize = 20)
plt.axis('off')

#---- 2nd plot ----
plt.subplot(1,5,2) # plot in 1 row and 5 columns, 2nd plot
size = df['race/ethnicity'].value_counts() # Here we are counting the values of race/ethnicity
labels = 'Group C', 'Group D','Group B','Group E','Group A'
color = ['red','green','blue','cyan','orange']

plt.pie(size, colors=color, labels=labels, autopct='%2f%%')
plt.title('Race/Ethnicity', fontsize = 20)
plt.axis('off')

#---- 3rd plot ----
plt.subplot(1,5,3) # plot in 1 row and 5 columns, 3rd plot
size=df['lunch'].value_counts() # here we are counting the values of lunch
labels = 'standard','Free'
color = ['red','green']

plt.pie(size, colors = color, labels = labels, autopct= '%2f%%') # here we are plotting pie chart for lunch with autopct to show percentage
plt.title('Lunch', fontsize = 20)
plt.axis('off')#  no axis will be shown

#---- 4th plot ----
plt.subplot(1,5,4) # plot in 1 row and 5 columns, 4th plot
size = df['test preparation course'].value_counts() # here we are counting the values of test preparation course
labels = 'none','completed'
color = ['red','green']

plt.pie(size,colors=color, labels=labels, autopct='%2f%%') # here we are plotting pie chart for test preparation course with autopct to show percentage
plt.title('Test Course', fontsize = 20)
plt.axis('off')# turn off the axis

# ---- 5th plot ----
plt.subplot(1,5,5) # plot in 1 row and 5 columns, 5th plot
size = df['parental level of education'].value_counts() # here we are counting the values of parental level of education
labels = ['some college','associate degree','high school','some high school','bachelor degree','master degree']
color = ['red','green','blue','cyan','orange','grey']

plt.pie(size, colors =color, labels=labels, autopct='%2f%%') # here we are plotting pie chart for parental level of education with autopct to show percentage
plt.title('Parental Education', fontsize = 20)
plt.axis('off')# turn off the axis

plt.tight_layout()
plt.grid()

plt.show()
```

#### BIVARIATE ANALYSIS ( Is gender has any impact on student's performance ? )

```python
gender_group = df.groupby('gender').mean(numeric_only=True)
gender_group
```

```python
plt.figure(figsize=(10,8))
X = ['Total Average Score', 'Math Average']
female_scores = [gender_group['average score'][0], gender_group['math score'][0]]
male_scores = [gender_group['average score'][1], gender_group['math score'][1]]

X_axis = np.arange(len(X))
plt.bar(X_axis - 0.2, male_scores, 0.4, label = 'Male')
plt.bar(X_axis + 0.2, female_scores, 0.4, label = 'Female')

plt.xticks(X_axis, X)
plt.ylabel("Marks")
plt.title("Total Average Score vs Math Average Score by Gender", fontweight = 'bold')
plt.legend()
plt.show()
```

#### Insights
-   On an average females have a better overall score than men.
